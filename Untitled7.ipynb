```python
import numpy as np
import pandas as pd

# Data klien A
data_a = np.random.randint(0, 50, 50)
# Data klien B
data_b = np.random.randint(0, 50, 50)

# Klien A menghitung rata-rata lokal
avg_a = np.mean(data_a)
print(f"Klien A rata-rata lokal: {avg_a}")

# Klien B menghitung rata-rata lokal
avg_b = np.mean(data_b)
print(f"Klien B rata-rata lokal: {avg_b}")

# Server menggabungkan (federated average)
federated_avg = (avg_a + avg_b) / 2
print(f"Federated Average: {federated_avg}")

# Visualisasi dengan matplotlib
import matplotlib.pyplot as plt

plt.hist([data_a, data_b], bins=20, label=['Klien A', 'Klien B'], alpha=0.7)
plt.axvline(avg_a, color='blue', linestyle='dashed', linewidth=1)
plt.axvline(avg_b, color='orange', linestyle='dashed', linewidth=1)
plt.axvline(federated_avg, color='green', linestyle='dashed', linewidth=2, label='Federated Avg')
plt.legend()
plt.title('Distribusi Data Klien A & B dengan Rata-rata')
plt.show()
```
